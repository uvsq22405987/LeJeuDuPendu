- création de fenetre de jeu
- creation du jeu pendu


```python
import tkinter as tk
from tkinter import *
import webbrowser
import random

## FONCTION POUR LA PAGE D'ACCUEIL ##
# Fonction pour ouvrir les règles du jeu
def open_rules():
    webbrowser.open_new("https://champagnole.circo39.ac-besancon.fr/wp-content/uploads/sites/9/2020/04/jeu-du-pendu.pdf") #webbrowser.open_new() permet d'ouvrir un lien dans le navigateur par défaut

# Fonction pour démarrer le jeu (à implémenter)
def start_game():
    print("Le jeu commence !")  # Remplace cette ligne par le code pour démarrer le jeu

#########################################

def jeu_pendu(mot_choisi) :
    global nb_erreurs, view_canvas #on met global pour pouvoir utiliser ces variables dans la fonction

    mot_cache = "*" * len(mot_choisi)  #on cache le mot avec des étoiles
    lettres_trouvees = []
    lettres_fausses = []
    nb_essais_max = 7
    nb_erreurs = 0

    while nb_erreurs < nb_essais_max : #tant que le joueur n'a pas perdu

        tentative = input("saisir une lettre").lower() #on demande au joueur de saisir une lettre et on la met en minuscule

        if not tentative.isalpha() or len(tentative) != 1 : #si la tentative n'est pas une lettre ou si elle n'est pas de longueur 1
            print("veuillez entrer qu'une seule lettre de l'alphabet") #on affiche un message d'erreur
            continue

        if (tentative in lettres_trouvees) or (tentative in lettres_fausses) : #si la lettre a déjà été essayée
            print("vous avez déjà essayé cette lettre") #on affiche un message d'erreur         
            continue

        if tentative in mot_choisi : #si la lettre est dans le mot à deviner
            lettres_trouvees.append(tentative) #on ajoute la lettre aux lettres trouvées
            mot_cache = "".join([tentative if tentative in lettres_trouvees else "*" for tentative in mot_choisi]) #on met les lettres trouvées dans le mot caché

        else :
            print(tentative, "vous vous êtes trompé") #on affiche un message d'erreur si la lettre n'est pas dans le mot
            lettres_fausses.append(tentative) #on ajoute la lettre aux lettres fausses
            print(lettres_fausses)
            nb_erreurs += 1 #on incrémente le nombre d'erreurs
            dessin_pendu(len(lettres_fausses)) #on affiche le dessin du pendu correspondant au nombre d'erreurs
            root.update() #on met à jour la fenêtre pour afficher le dessin du pendu 

        print(mot_cache)  ##mis à la fin, après avoir traité la lettre (la fameuse tentative), car comme ça le joueur peut voir l'avancée de son mot au fur et à mesure.
        nb_erreurs = len(lettres_fausses) #on met à jour le nombre d'erreurs en fonction du nombre de lettres fausses

        if mot_cache == mot_choisi : #si le mot caché est égal au mot choisi, c'est-à-dire si le joueur a trouvé toutes les lettres
            print("c'est gagné !")
            break #on sort de la boucle

    if mot_cache != mot_choisi : #si le mot caché n'est pas égal au mot choisi, c'est-à-dire si le joueur n'a pas trouvé toutes les lettres
        print("vous avez atteint le nombre maximum d'essais. Le mot à deviner était", mot_choisi) #on affiche le mot à deviner

#############################################

##CREATION DES CANVAS
def page_accueil():
    # Création de la fenêtre principale
    window = Tk() #fenêtre principale
    window.title("JEU DU PENDU") #titre de la fenêtre
    window.geometry("1080x720") #taille de la fenêtre
    window.config(bg='#2E3440')  # Fond sombre et moderne

    # Bouton "Aide" en haut à gauche (orange)
    button_help = Button(
        window, 
        text="Aide", 
        font=("Helvetica", 14, "bold"), 
        bg='#FFA500',  # Orange
        fg='white', 
        activebackground='#FF8C00',  # Orange plus foncé au survol
        activeforeground='white', 
        relief=FLAT, 
        bd=0, 
        padx=15, 
        pady=10, 
        command=open_rules
    ) #bouton aide qui ouvre les règles du jeu dans un navigateur web                                                                                                           
    button_help.place(x=20, y=20)  # Position en haut à gauche

    # Frame principale
    frame = Frame(window, bg='#2E3440', bd=1, relief=SUNKEN) #cadre principal de la fenêtre 
    frame.place(relx=0.5, rely=0.5, anchor=CENTER) #position du cadre au centre de la fenêtre 

    # Premier titre
    label_titre = Label(
        frame, 
        text="BIENVENUE AU JEU DU PENDU", 
        font=("Helvetica", 30, "bold"), 
        bg='#2E3440', 
        fg='#88C0D0'  # Couleur pastel pour le texte
    ) #titre du jeu 
    label_titre.grid(row=0, column=0, pady=20, padx=20) #position du titre au centre du cadre 

    # Label "Fait par nous"
    label_subtitre = Label(
        frame, 
        text="Fait par Klervi,Mathis,Chelsea,Ali", 
        font=("Helvetica", 15, "italic"), 
        bg='#2E3440', 
        fg='#D8DEE9'  # Texte gris clair
    ) #label sous-titre du jeu 
    label_subtitre.grid(row=1, column=0, pady=10) #position du sous-titre au centre du cadre

    # Bouton "Jouer" avec style moderne
    button_play = Button(
        frame, 
        text="Jouer", 
        font=("Helvetica", 20, "bold"), 
        bg='#5E81AC',  # Bleu moderne
        fg='white', 
        activebackground='#81A1C1',  # Bleu plus clair au survol
        activeforeground='white', 
        relief=FLAT, 
        bd=0, 
        padx=20, 
        pady=10, 
        command=start_game
    ) #bouton jouer qui lance le jeu
    button_play.grid(row=2, column=0, pady=20) #position du bouton jouer au centre du cadre

    # Ajouter un effet de survol au bouton "Aide"
    def on_enter_help(e): #fonction qui s'active quand le bouton est survolé
        button_help.config(bg='#FF8C00')  # Orange plus foncé au survol

    def on_leave_help(e): #fonction qui s'active quand le bouton n'est plus survolé
        button_help.config(bg='#FFA500')  # Retour à l'orange d'origine

    button_help.bind("<Enter>", on_enter_help) #ajouter l'effet de survol au bouton "Aide"
    button_help.bind("<Leave>", on_leave_help) #retirer l'effet de survol au bouton "Aide"

    # Ajouter un effet de survol au bouton "Jouer"
    def on_enter_play(e): #fonction qui s'active quand le bouton est survolé
        button_play.config(bg='#81A1C1')  # Bleu plus clair au survol

    def on_leave_play(e): #fonction qui s'active quand le bouton n'est plus survolé
        button_play.config(bg='#5E81AC')  # Retour au bleu d'origine

    button_play.bind("<Enter>", on_enter_play) #ajouter l'effet de survol au bouton "Jouer"
    button_play.bind("<Leave>", on_leave_play) #retirer l'effet de survol au bouton "Jouer"

    # Lancer la boucle Tkinter
    window.mainloop() #boucle principale du jeu
```
